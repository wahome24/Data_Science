```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the iris dataset from seaborn
df= sns.load_dataset('iris')
```

```python
#Dataset stats
df.describe()
```

```python
#Nummber of entries for each species
df['species'].value_counts()
```

```python
#Visualizing the dataset variables

sns.pairplot(df, diag_kind='kde',hue='species')
```

```python
#Adding linear regression line
sns.pairplot(df,kind='reg',hue='species')
```

```python
#Pairplot with defined columns to display
sns.pairplot(df, diag_kind='kde',vars=['sepal_width','sepal_length'],hue='species')
```
