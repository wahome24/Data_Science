### Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\USER\Desktop\Station Data\Stations.csv")
```

```python
df
```

```python
#The first step is to check if there is duplicated data.

df.drop_duplicates()
```

```python
#This step is to remove the To be del column

df.drop(columns = "To be del")
```

```python
#Updating the entire data frame to refelect the dropped column

df = df.drop(columns = "To be del")
df
```

```python
#The next step will be to replace the missing values marked by --- with blanks.

df.replace("---","")

df = df.replace("---","")

df
```

```python
#The next step is to convert the blanks to NaN usiing below regex expression.

df.replace(r'^\s*$', float("NaN"),regex=True)
df = df.replace(r'^\s*$', float("NaN"),regex=True)
df
```

```python
#Drop NaN

df.dropna()
df = df.dropna()
df
```

```python
#The final step is to remove the #, $ and * on some of the row data.
#I am using a regular expression that identifies characters that are not letters or numbers without removing - or .

df.replace(r'[^\w\.\-]',"",regex=True)
df = df.replace(r'[^\w\.\-]',"",regex=True)
df
```

```python
#Resetting index values

df.reset_index(drop=True)
df = df.reset_index(drop=True)
df
```

### Classification

```python
df.dtypes #checking for the column datatypes
```

```python
#Will need to convert the object datatypes into float apart from station and with af days, mm and yyyy being int    

df[["tmax degC","tmin degC","rain mm","sun hours"]] = df[["tmax degC","tmin degC","rain mm","sun hours"]].astype(float)
```

```python
#convert af days into int

df["af days"] = df["af days"].astype(int)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
#Importing libraries to be used

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
#Function to create categories based on latitude

def classify_latitude(lat):
    if lat > 57.3:
        return 'Northern Third UK'
    elif lat >= 53.6 and lat <= 57.2:
        return 'Central Third UK'
    else:
        return 'Southern Third UK'
```

```python
#Update the data frame to reflect the categories above

df['Categories'] = df['Latitude'].apply(classify_latitude)
df
```

```python
# Features and Labels
X = df[['Latitude', 'Longitude']]
y = df['Categories']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Initialize the Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=50)
```

```python
# Train the classifier
clf.fit(X_train, y_train)
```

```python
# Predict the classifications and print classification report
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```
