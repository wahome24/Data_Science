```python
import sklearn as sk
import sklearn.datasets as skd
import sklearn.cluster as skc
```

```python
import matplotlib.pyplot as plt
```

```python
X, y = skd.make_blobs()
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
model = skc.KMeans(n_clusters=3)
```

```python
plt.scatter(X[:,0],X[:,1],c=model.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```
