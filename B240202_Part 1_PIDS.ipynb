### Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\USER\Desktop\Station Data\Station Data.csv")
```

```python
df
```

```python
#The first step is to check if there is duplicated data.

df.drop_duplicates()
```

```python
#This step is to remove the To be del column

df.drop(columns = "To be del")
```

```python
#Updating the entire data frame to refelect the dropped column

df = df.drop(columns = "To be del")
df
```

```python
#The next step will be to replace the missing values marked by --- with blanks.

df.replace("---","")

df = df.replace("---","")

df
```

```python
#The next step is to convert the blanks to NaN usiing below regex expression.

df.replace(r'^\s*$', float("NaN"),regex=True)
df = df.replace(r'^\s*$', float("NaN"),regex=True)
df
```

```python
#Drop NaN

df.dropna()
df = df.dropna()
df
```

```python
#The final step is to remove the #, $ and * on some of the row data.
#I am using a regular expression that identifies on characters that are not letters or numbers

df.replace(r'[^\w\.\-]',"",regex=True)
df = df.replace(r'[^\w\.\-]',"",regex=True)
df
```

```python
#Resetting index values

df.reset_index(drop=True)
df = df.reset_index(drop=True)
df
```

```python
#Exporting the cleaned data

df.to_csv("C:\\Users\\USER\\Desktop\\Station Data\\Station_Data_Cleaned.csv")
```

### Clustering

```python
df.dtypes #checking for the column datatypes
```

```python
#Will need to convert the object datatypes into float apart from station  and with af days, mm and yyyy being int    

df[["tmax degC","tmin degC","rain mm","sun hours"]] = df[["tmax degC","tmin degC","rain mm","sun hours"]].astype(float)
```

```python
#convert af days into int

df["af days"] = df["af days"].astype(int)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
#Removing the station name column to remain only with the interested variables

df = df.drop(columns="Station")
df
```

```python
#Importing libraries to be used
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# transforming the data using StandardScaler

scaler = StandardScaler()
```

```python
#Creating standardized copies of tmax degC, tmin degC,af days,rain mm,sun hours 

df[["tmax","tmin","af_days","rain","sun"]] = scaler.fit_transform(df[["tmax degC","tmin degC","af days","rain mm","sun hours"]])
```

```python
# Original Data Frame including transformed rows
df
```

```python
#Function to determine the required number of clusters

def cluster(data, max_k):
    means = []
    inertias = []
    
    for k in range(1,max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        
        means.append(k)
        inertias.append(kmeans.inertia_)
        
        #Generating the Elbow Plot
        fig = plt.subplots(figsize=(10,5))
        plt.plot(means,inertias,'o-')
        plt.xlabel('Num of Clusters')
        plt.ylabel('Inertia')
        plt.grid(True)
        plt.show()
```

```python
#Calling the function and providing data and maximum number of clusters required

cluster(df[["tmax","tmin"]],10)
```

```python
#KMeans clustering application on intereted columns
# Chosen mumber of clusters = 4

kmeans = KMeans(n_clusters=4)
kmeans.fit(df[["tmax","tmin","af_days","rain","sun"]])
```

```python
df["Clusters"] = kmeans.labels_
df
```

```python
#Plotting for min temp againt max temp
plt.scatter(x=df["tmin"],y=df["tmax"],c=df["Clusters"])
plt.xlabel('Min Temp')
plt.ylabel('Max Temp')
plt.show()
```

```python
#Plotting for total sunshine against rain
plt.scatter(x=df["sun"],y=df["rain"],c=df["Clusters"])
plt.xlabel('Total Sunshine')
plt.ylabel('Total Rainfall')
plt.show()
```
