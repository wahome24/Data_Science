```python
#Load hapiness data

happy = pd.read_csv(r"C:\Users\USER\Desktop\Station Data\Happiness_Data.csv")
```

```python
happy
```

```python
#Removing unwanted columns from the weather data

happy = happy.drop(columns=["Area Code","Region","Region Code"])

happy
```

```python
#Function to create categories based on latitude

def classify_latitude(lat):
    if lat > 54.4:
        return 'Northern Third UK'
    elif lat >= 52.7 and lat <= 54.3:
        return 'Central Third UK'
    else:
        return 'Southern Third UK'
```

```python
#Update the data frame to reflect the categories above

happy['Categories'] = happy['Latitude'].apply(classify_latitude)
happy
```

```python
#Based on the categories for the happiness data we shall update the averages for the weather data
#Functions to insert averages for the weather data

def tmax_degC(tmax):
    if tmax == 'Northern Third UK':
        return 10.89
    elif tmax == 'Central Third UK':
        return 12.12
    else:
        return 13.77
    
def tmin_degC(tmin):
    if tmin == 'Northern Third UK':
        return 5.25
    elif tmin == 'Central Third UK':
        return 5.29
    else:
        return 6.68 
    
def rain_mm(rain):
    if rain == 'Northern Third UK':
        return 82.89
    elif rain == 'Central Third UK':
        return 82.72
    else:
        return 65.53
    
def sun_hours(sun):
    if sun == 'Northern Third UK':
        return 99.37
    elif sun == 'Central Third UK':
        return 108.17
    else:
        return 127.70
    
```

```python
#Updating the data frame to reflect the new columns of weather data

happy['tmax degC'] = happy['Categories'].apply(tmax_degC)
happy['tmin degC'] = happy['Categories'].apply(tmin_degC)
happy['rain mm'] = happy['Categories'].apply(rain_mm)
happy['sun hours'] = happy['Categories'].apply(sun_hours)
```

```python
happy.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#Plotting for Mean daily max temperature againt happiness rating

plt.scatter(x=happy["tmax degC"],y=happy["Hapiness Rating"])
plt.xlabel("Mean daily max temperature")
plt.ylabel("Happiness Rating")
plt.show()
```

```python
#Calculating the averages for the happiness data based on categories

happy.groupby("Categories")["Hapiness Rating"].mean()
```

```python
def classify_happiness(hap):
    if hap == 'Northern Third UK':
        return 7.53
    elif hap == 'Central Third UK':
        return 7.48
    else:
        return 7.50
```

```python
#Updating the data frame to reflect the new column of happiness data

df['Happiness_Rating'] = df['Categories'].apply( classify_happiness)
```

```python
#Plotting for Mean daily min temperature againt total rainfall

plt.scatter(df["tmin degC"],df["rain mm"],c=df["Happiness_Rating"])
plt.xlabel("Mean daily min temperature")
plt.ylabel("Total Rainfall")
plt.show()
```
